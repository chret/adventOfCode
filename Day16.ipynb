```python
basePattern=[0, 1, 0, -1]

def fftOneStep(inputSeq,basePattern):
    outputSeq=[0]*len(inputSeq)
    for i in range(len(inputSeq)):
        print(i)
        curBasePattern=[]
        while len(curBasePattern) < len(inputSeq)+1:
            for b in range(len(basePattern)):
                curBasePattern += [basePattern[b]]*(i+1)
                if len(curBasePattern) >= len(inputSeq)+1:
                    break
        print(i)
        val = sum(e[0]*e[1] for e in zip(inputSeq[i:],curBasePattern[i+1:len(inputSeq)+1]))
        #for j in range(i,len(inputSeq)):
        #    val += inputSeq[j] * curBasePattern[j+1]
        outputSeq[i] = abs(val)%10
        print(i)
        
    return outputSeq
        
    
def fastfftOneStep(inputSeq,basePattern):
    outputSeq=[0]*len(inputSeq)
    for i in range(0,len(inputSeq)):
        #val=sum(inputSeq[j] * basePattern[((j+1)//(i+1))%4] for j in range(i,len(inputSeq)))
        val=0
        for j in range(i,len(inputSeq)):
            val += inputSeq[j] * basePattern[((j+1)//(i+1))%4]
        #val=sum(inputSeq[i:])
        outputSeq[i] = abs(val)%10
        #print(i)
        
    return outputSeq

# this uses the fact that only indices larger than I matter for the computation of element I in the output sequence
# so if we have a large offset, we can directly start there
# also, I saw that with such a large offset, the basePattern reduces to '1's, i.e. a simple sum
# this also holds for all elements in range (offset,END)
# instead of recalcaulting the sum of all remainder elements, 
# I reuse the sum of the previous one and subtract the summand that was too much, i.e. the previous one
# care has to be taken with modulo arithmetic, so this has to be done afterwards
# TLDR: it's ugly
def cheatedfastfftOneStep(inputSeq,basePattern,offset):
    outputSeq=[0]*len(inputSeq)
    outputSeq[offset] = sum(inputSeq[offset:])
    for i in range(offset+1,len(inputSeq)):
        val=outputSeq[i-1]-inputSeq[i-1]
        outputSeq[i] = val#abs(val)%10
        #print(i)
    for i in range(offset+1,len(inputSeq)):
        outputSeq[i] = abs(outputSeq[i])%10
    
    return outputSeq
```
