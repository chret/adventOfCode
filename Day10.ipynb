```python
import numpy as np
file=open("input_day10.txt","r")
fileLines = file.read().splitlines()
#print(fileLines)
file.close()
```

```python
asteroidMap = np.array([np.array([i == '#' for i in line]) for line in fileLines])
monitoringNumbermap = np.zeros(asteroidMap.shape)
print(asteroidMap)

for y in range(0,asteroidMap.shape[0]):
    for x in range(0,asteroidMap.shape[1]):
        if asteroidMap[y][x]:
            numberOfMonitors = 0
            blockedAngles=[]
            for yy in range(0,asteroidMap.shape[0]):
                for xx in range(0,asteroidMap.shape[1]):
                    if xx == x and yy == y:
                        continue
                    if asteroidMap[yy][xx]:
                        angle = np.arctan2(float(yy-y),float(xx-x))
                        if angle in blockedAngles:
                            continue
                            # todo float compare?
                        else:
                            numberOfMonitors +=1
                            blockedAngles.append(angle)
            monitoringNumbermap[y][x] = numberOfMonitors
```

```python
print(monitoringNumbermap)
```

```python
print(np.amax(monitoringNumbermap))
```

```python
coords = np.where(monitoringNumbermap == np.amax(monitoringNumbermap))
yx = list(zip(coords[0], coords[1]))[0]
```

```python
print(yx)
print(monitoringNumbermap[yx[0]][yx[1]])
print(np.arctan2(-1,-1))
```

```python
angleDistCoordList=[]
x=yx[1]
y=yx[0]
for yy in range(0,asteroidMap.shape[0]):
    for xx in range(0,asteroidMap.shape[1]):
        if xx == x and yy == y:
            continue
        if asteroidMap[yy][xx]:
            angle = np.arctan2(float(xx-x),float(y-yy)) # y direction is inverted from "regular" coordinate system
            if angle < 0:
                angle += 2*np.pi
            dist = np.sqrt((xx-x)**2+(yy-y)**2)
            angleDistCoordList.append((angle,dist,(xx,yy)))
```

```python
#print(angleDistCoordList)

#sort by first angle and then distance
from operator import itemgetter
angleDistCoordList.sort(key=itemgetter(0,1))

print(len(angleDistCoordList),angleDistCoordList)
# hyptoteiccally delete 200 asteroids from list
curAngle=-1
index=0
for i in range(0,len(angleDistCoordList)):
    if angleDistCoordList[i][0] > curAngle:
        curAngle = angleDistCoordList[i][0]
        index += 1
        if index == 200:
            print(i,angleDistCoordList[i])
```
