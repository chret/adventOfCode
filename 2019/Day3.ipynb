```python
import numpy as np
```

```python
ws=np.loadtxt("input_day3.txt",delimiter=",",dtype=str,skiprows=0)
w1 = ws[0]
w2 = ws[1]
```

```python
print(w1,w2)
```

```python
def createCoordListFromInput(wire):
    wPoints=[(0,0)]
    for i,entry in enumerate(wire):
        direction=entry[0]
        steps=int(entry[1:])
        oldPoint = wPoints[i]
        if direction=="R":
            wPoints.append((oldPoint[0]+steps,oldPoint[1]))
        elif direction=="L":
            wPoints.append((oldPoint[0]-steps,oldPoint[1]))
        elif direction=="U":
            wPoints.append((oldPoint[0],oldPoint[1]+steps))
        elif direction=="D":
            wPoints.append((oldPoint[0],oldPoint[1]-steps))
    return wPoints        
```

```python
w1Points=createCoordListFromInput(w1)
w2Points=createCoordListFromInput(w2)
print(w1Points)
```

```python
minimalDist = 1000000000000000
fewestCombinedSteps=1000000000000
steps1=0
for i in range(0,len(w1Points)-1):
    start1=w1Points[i]
    end1=w1Points[i+1]
    delta1=(end1[0]-start1[0],end1[1]-start1[1])
    steps2=0
    for j in range(0,len(w2Points)-1):
        start2=w2Points[j]
        end2=w2Points[j+1]
        delta2=(end2[0]-start2[0],end2[1]-start2[1])
        
        #print("------")
        #print(start1, end1, delta1,steps1)
        #print(start2, end2, delta2,steps2)
            
        if delta1[1] == 0 and delta2[0] == 0:
            # wire 1 parallel to x
            int1 = [start1[0],end1[0]]
            int2 = [start2[1],end2[1]]
            int1.sort()
            int2.sort()
            
            if (int1[0] <= start2[0] and start2[0] <= int1[1] and int2[0] <= start1[1] and start1[1] <= int2[1]):
                print("Found intersection point", start2[0], start1[1])
                dist = abs(start2[0])+ abs(start1[1])
                if dist > 0:
                    minimalDist = min(dist,minimalDist)
                    combinedSteps = steps1 + abs(start2[0]-start1[0]) + steps2 + abs(start1[1]-start2[1])
                    fewestCombinedSteps=min(combinedSteps, fewestCombinedSteps)
                
        if delta1[0] == 0 and delta2[1] == 0:
            # wire 1 parallel to x
            int1 = [start1[1],end1[1]]
            int2 = [start2[0],end2[0]]
            int1.sort()
            int2.sort()
            
            if (int1[0] <= start2[1] and start2[1] <= int1[1] and int2[0] <= start1[0] and start1[0] <= int2[1]):
                print("Found intersection point",start1[0], start2[1])
                dist = abs(start1[0])+ abs(start2[1])
                if dist > 0:
                    minimalDist = min(dist,minimalDist)
                    combinedSteps = steps1 + abs(start2[1]-start1[1]) + steps2 + abs(start1[0]-start2[0])
                    fewestCombinedSteps=min(combinedSteps, fewestCombinedSteps)
                    
        
        steps2 = steps2 + abs(delta2[0]) + abs(delta2[1])
        
    steps1 = steps1 + abs(delta1[0]) + abs(delta1[1])
        
```
