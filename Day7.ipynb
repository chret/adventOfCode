```python
def decodeCommand(command):
    code=command%100
    p1Mode=int(command/100)%10
    p2Mode=int(command/1000)%10
    p3Mode=int(command/10000)%10
    return [code, p1Mode, p2Mode, p3Mode]

def getArgument(opcode, mode, parameter):
    if mode == 0:
        # position mode
        return opcode[parameter]
    else:
        # immediate mode
        return parameter
```

```python
def processOpcode(opcode, idx):
    
    decoded = decodeCommand(opcode[idx])
    
    if decoded[0] == 99:
        return
    elif decoded[0] == 1:
        #add p1 p2 p3
        p1 = opcode[idx+1]
        p2 = opcode[idx+2]
        p3 = opcode[idx+3]
        arg1 = getArgument(opcode,decoded[1],p1)
        arg2 = getArgument(opcode,decoded[2],p2)
        opcode[p3] =  arg1 + arg2
        return processOpcode(opcode,idx+4)
    elif decoded[0] == 2:
        #multiply p1 p2 p3
        p1 = opcode[idx+1]
        p2 = opcode[idx+2]
        p3 = opcode[idx+3]
        arg1 = getArgument(opcode,decoded[1],p1)
        arg2 = getArgument(opcode,decoded[2],p2)
        opcode[p3] =  arg1 * arg2
        return processOpcode(opcode,idx+4)
    elif decoded[0] == 3:
        #input p1
        p1 = opcode[idx+1]
        var=0
        if len(inputList)>0:
            var = inputList.pop(0)
        else:
            var = input("InputList is empty, please enter input manually: ")
        opcode[p1] = int(var)
        return processOpcode(opcode,idx+2)
    elif decoded[0] == 4:
        #output p1
        p1 = opcode[idx+1]
        arg1 = getArgument(opcode,decoded[1],p1)
        print(arg1)
        outputList.append(arg1)
        return processOpcode(opcode,idx+2)
    elif decoded[0] == 5:
        #jump-if-true p1 p2
        p1 = opcode[idx+1]
        p2 = opcode[idx+2]
        arg1 = getArgument(opcode,decoded[1],p1)
        arg2 = getArgument(opcode,decoded[2],p2)
        if arg1 != 0:
            return processOpcode(opcode,arg2)
        else:      
            return processOpcode(opcode,idx+3)
    elif decoded[0] == 6:
        #jump-if-false p1 p2
        p1 = opcode[idx+1]
        p2 = opcode[idx+2]
        arg1 = getArgument(opcode,decoded[1],p1)
        arg2 = getArgument(opcode,decoded[2],p2)
        if arg1 == 0:
            return processOpcode(opcode,arg2)
        else:      
            return processOpcode(opcode,idx+3)
    elif decoded[0] == 7:
        #less-than p1 p2 p3
        p1 = opcode[idx+1]
        p2 = opcode[idx+2]
        p3 = opcode[idx+3]
        arg1 = getArgument(opcode,decoded[1],p1)
        arg2 = getArgument(opcode,decoded[2],p2)
        if arg1 < arg2:
            opcode[p3]=1
        else:      
            opcode[p3]=0
        return processOpcode(opcode,idx+4)
    elif decoded[0] == 8:
        #equals p1 p2 p3
        p1 = opcode[idx+1]
        p2 = opcode[idx+2]
        p3 = opcode[idx+3]
        arg1 = getArgument(opcode,decoded[1],p1)
        arg2 = getArgument(opcode,decoded[2],p2)
        if arg1 == arg2:
            opcode[p3]=1
        else:      
            opcode[p3]=0
        return processOpcode(opcode,idx+4)
    else:
        print("Invalid code ", opcode, idx, opcode[idx], decoded)
        return
```

```python
import threading, queue

class WorkerThread(threading.Thread):
    def __init__(self, opcode, in_q, out_q, number):
        super(WorkerThread, self).__init__()
        self.opcode = opcode
        self.in_q = in_q
        self.out_q = out_q
        self.number = number

    def run(self):
        self.processOpcode(self.opcode,0)  
        #print("Thread " + str(self.number) + " terminated run") 
    
    def processOpcode(self,opcode,idx):
    
        decoded = decodeCommand(opcode[idx])

        if decoded[0] == 99:
            print("Thread " + str(self.number) + " finishing")
            return
        elif decoded[0] == 1:
            #add p1 p2 p3
            p1 = opcode[idx+1]
            p2 = opcode[idx+2]
            p3 = opcode[idx+3]
            arg1 = getArgument(opcode,decoded[1],p1)
            arg2 = getArgument(opcode,decoded[2],p2)
            opcode[p3] =  arg1 + arg2
            return self.processOpcode(opcode,idx+4)
        elif decoded[0] == 2:
            #multiply p1 p2 p3
            p1 = opcode[idx+1]
            p2 = opcode[idx+2]
            p3 = opcode[idx+3]
            arg1 = getArgument(opcode,decoded[1],p1)
            arg2 = getArgument(opcode,decoded[2],p2)
            opcode[p3] =  arg1 * arg2
            return self.processOpcode(opcode,idx+4)
        elif decoded[0] == 3:
            #input p1
            p1 = opcode[idx+1]
            #print("Thread " + str(self.number) + " waiting for input")
            var = self.in_q.get(True) # this will block until entry is found
            #print("Thread " + str(self.number) + " continuing with input " + str(var))
            opcode[p1] = int(var)
            return self.processOpcode(opcode,idx+2)
        elif decoded[0] == 4:
            #output p1
            p1 = opcode[idx+1]
            arg1 = getArgument(opcode,decoded[1],p1)
            print(arg1)
            self.out_q.put(arg1)
            return self.processOpcode(opcode,idx+2)
        elif decoded[0] == 5:
            #jump-if-true p1 p2
            p1 = opcode[idx+1]
            p2 = opcode[idx+2]
            arg1 = getArgument(opcode,decoded[1],p1)
            arg2 = getArgument(opcode,decoded[2],p2)
            if arg1 != 0:
                return self.processOpcode(opcode,arg2)
            else:      
                return self.processOpcode(opcode,idx+3)
        elif decoded[0] == 6:
            #jump-if-false p1 p2
            p1 = opcode[idx+1]
            p2 = opcode[idx+2]
            arg1 = getArgument(opcode,decoded[1],p1)
            arg2 = getArgument(opcode,decoded[2],p2)
            if arg1 == 0:
                return self.processOpcode(opcode,arg2)
            else:      
                return self.processOpcode(opcode,idx+3)
        elif decoded[0] == 7:
            #less-than p1 p2 p3
            p1 = opcode[idx+1]
            p2 = opcode[idx+2]
            p3 = opcode[idx+3]
            arg1 = getArgument(opcode,decoded[1],p1)
            arg2 = getArgument(opcode,decoded[2],p2)
            if arg1 < arg2:
                opcode[p3]=1
            else:      
                opcode[p3]=0
            return self.processOpcode(opcode,idx+4)
        elif decoded[0] == 8:
            #equals p1 p2 p3
            p1 = opcode[idx+1]
            p2 = opcode[idx+2]
            p3 = opcode[idx+3]
            arg1 = getArgument(opcode,decoded[1],p1)
            arg2 = getArgument(opcode,decoded[2],p2)
            if arg1 == arg2:
                opcode[p3]=1
            else:      
                opcode[p3]=0
            return self.processOpcode(opcode,idx+4)
        else:
            print("Invalid code ", opcode, idx, opcode[idx], decoded)
            return

    def join(self, timeout=None):
        super(WorkerThread, self).join(timeout)
```

```python
import itertools   
phaseOptions = range(5,10)
perm = itertools.permutations(phaseOptions)
maxThrust=0
for entry in perm:
    queues = [queue.Queue() for i in range(0,5)]

    # Create the "thread pool"
    pool = [WorkerThread(opcode=opcodeInput.copy(), in_q=queues[i], out_q=queues[(i+1)%5],number=i) for i in range(0,5)]

    # Start all threads
    for thread in pool:
        thread.start()

    for worker,phase in enumerate(entry):
        queues[worker].put(phase)
    queues[0].put(0)
    
    # Ask threads to die and wait for them to do it
    for thread in pool:
        thread.join()
    
    maxThrust = max(maxThrust,queues[0].get())
```
