```python
file=open("input_day6.txt","r")
fileLines=file.readlines()

print(fileLines)

file.close()
```

```python
def getSrcAndDst(fileLine):
    idx = fileLine.find(")")
    src = fileLine[0:idx]
    dst = fileLine[(idx+1):fileLine.find("\n")]
    return (src,dst)
```

```python
data = [("COM", "COM", 0)]
dataPoints = len(fileLines)
insertedDataPoints=0
currentOrbitSources=["COM"]
nextOrbitSources=[]
currentOrbitLevel=1
totalOrbitLinks=0
while(len(data) < dataPoints+1):
    for line in fileLines:
        entry = getSrcAndDst(line)
        if entry[0] in currentOrbitSources:
            data.append((entry[1],entry[0],currentOrbitLevel))
            nextOrbitSources.append(entry[1])
            totalOrbitLinks += currentOrbitLevel
    currentOrbitLevel += 1
    currentOrbitSources = nextOrbitSources.copy()
    nextOrbitSources = []
print(data)
```

```python
print(totalOrbitLinks)
```

```python
def wayFromOrbitToCOM(orbit, data):
    wayToCom=[]
    currentOrbit=orbit
    while(currentOrbit != "COM"):
        for o, parent, l in data:
            if o == currentOrbit:
                currentOrbit = parent
                wayToCom.append(parent)
    return wayToCom
```

```python
wayFromMeToCOM = wayFromOrbitToCOM("YOU", data)
wayFromSantaToCOM = wayFromOrbitToCOM("SAN", data)
```

```python
wayFromMeToCOM.reverse()
wayFromSantaToCOM.reverse()

lastCommonOrbit="COM"
for entry in wayFromSantaToCOM:
    if entry in wayFromMeToCOM:
        lastCommonOrbit = entry
    else:
        break
```

```python
linksToLastCommonOrbit = 0
for x,y,z in data:
    if x == lastCommonOrbit:
        linksToLastCommonOrbit = z
print(linksToLastCommonOrbit)

linksToMe = 0
for x,y,z in data:
    if x == "YOU":
        linksToMe = z
print(linksToMe)

linksToSanta = 0
for x,y,z in data:
    if x == "SAN":
        linksToSanta = z
print(linksToSanta)
```

```python
linksFromMeToSanta = (linksToMe-linksToLastCommonOrbit-1) + (linksToSanta-linksToLastCommonOrbit-1)
print(linksFromMeToSanta)
```
